# NPV Summary Table

```python
series_order = ['ERCOT', 'CAISO', 'NYISO', 'MISO-W', 'PJM-W', 'BaseCaseTax', 'HighWindTax', 'HighSolarTax', 'WinterNYTax']
```

```python
def output_NPV_excel(file, output_name):
    df = pd.read_csv(file)
    df['scenario'] = df['scenario'].apply(lambda x: x if '_' not in x else x.split('_')[-1])

    df_NPV_table = pd.DataFrame()

    for _,row in df.iterrows():
        df_NPV_table.loc[row['CO2_price'], row['scenario']] = row['NPV']

    df_NPV_table.sort_index(inplace=True)
    df_NPV_table = df_NPV_table[series_order]
    df_NPV_table.to_excel(output_name)
```

```python
file_names = ['START-5-NG-383_NPVs_w_B31A_PCC_25',
              'START-5-NG-383_NPVs_w_B31A_PCC_40',
              'START-5-NG-383_NPVs_w_B31A',
              'START-5-NG-450_NPVs_w_B31A',
              'START-5-NG-600_NPVs_w_B31A',
              ]

for file_name in file_names:
    file = '../../../results/revamp-start-up/' + file_name + '.csv'
    output = '../../../results/revamp-start-up/' + file_name + '.xlsx'

    output_NPV_excel(file, output)
```
