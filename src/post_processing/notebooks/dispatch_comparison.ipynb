```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading dispatch solution

```python
n_month = 12
CO2_prices = [100, 150, 225, 300]
prefix = 'START-5-NG-383-'

dispatch_dfs = {}

for CO2_price in CO2_prices:

    if CO2_price == 100:
        price_signal = 'MiNg_100_ERCOT'
    else:
        price_signal = 'MiNg_150_ERCOT'

    folder_path = '/Users/pengfeicheng/Library/CloudStorage/OneDrive-GeorgiaInstituteofTechnology/FLECCS/NPV-update/' + \
        prefix + str(n_month) + '-month-' + str(CO2_price) + '-' + price_signal + '/'

    dispatch_dfs[CO2_price] = pd.read_csv(folder_path + 'results_power.csv')['x_load_factor']
```

## Read price signal

```python
df_price = pd.read_csv("/Users/pengfeicheng/Documents/research/FLECCS/code/src/resources/overall-price-signals.csv")
```

## Dispatch + price

```python
end_h = 600
```

```python
fig, axes = plt.subplots(2, 2, figsize=(26, 15))

# dispatch
for j in range(2):

    i1, i2 = j // 2, j % 2
    ax = axes[i1, i2]
    ax.plot(dispatch_dfs[CO2_prices[j]][:end_h], 'b-', linewidth=1)
    ax.set_xlim([0, end_h])
    ax.set_ylim([-5, 105])
    ax.set_title(f"({['a', 'b', 'c', 'd'][j]}) CO2 price = {CO2_prices[j]}", fontsize=40, pad=15)
    plt.setp(ax.get_xticklabels(), Fontsize=20)
    plt.setp(ax.get_yticklabels(), Fontsize=20)

    # if j == 3:
    ax.set_xlabel("hour", fontsize=30)
    ax.set_ylabel("load factor", fontsize=30)

# price
for j in [2, 3]:

    if j == 2:
        price_signal = 'MiNg_100_ERCOT'
    else:  # 4
        price_signal = 'MiNg_150_ERCOT'

    i1, i2 = j // 2, j % 2
    ax = axes[i1, i2]

    price = df_price[price_signal]

    ax.plot(price[:end_h], 'r-', label='load factor')

    plt.setp(ax.get_xticklabels(), Fontsize=20)
    plt.setp(ax.get_yticklabels(), Fontsize=20)
    ax.set_ylabel("electricity price/$/MWh", fontsize=30)
    ax.set_xlabel("hour", fontsize=30)
    ax.set_xlim([0, end_h])
    ax.set_ylim([-5, 130])


plt.tight_layout(pad=3)
```
